```python
#DMSO for columns (A-F)
col_number = 1
for col in used_cols:
    pipette.transfer(
        dmso_vol1,
        tubes1.wells(dmso_loc),
        plate.cols(col),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
    col_number += 1
```

```python
#DMSO for rows (1-6)
for row in used_rows2:
        pipette.transfer(
        dmso_vol2,
        tubes1.wells(dmso_loc),
        plate.rows(row),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
```

```python
#DMSO for rows (6-12)
for row in used_rows3:
        pipette.transfer(
        dmso_vol3,
        tubes1.wells(dmso_loc),
        plate.rows(row),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
```

```python
#Drug for Columns (A-F)
col_number = 1
for col in used_cols:
    pipette.transfer(
        drug_vol1,
        tubes1.wells(drug_stock_loc[col]),
        plate.cols(col),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
    col_number += 1
```

```python
#Drug for Rows (1-6)
for row in used_rows2:
        pipette.transfer(
        drug_vol2,
        tubes1.wells(drug_stock_loc[row]),
        plate.rows(row),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
```

```python
#Drug for Rows (6-12)
for row in used_rows3:
        pipette.transfer(
        drug_vol3,
        tubes1.wells(drug_stock_loc[row]),
        plate.rows(row),
        blow_out=True,
        new_tip='always'
        touch_tip=True)
```
